```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("1000Companies.csv")
```

```python
df
```

```python
print(df.columns)
```

```python
summary_stats = df.describe()
print(summary_stats)
```

```python
newdata = df
```

```python
newdata
```

```python
newdata.columns = newdata.columns.str.strip()
```

```python
cols_to_clean = ['revenues', 'profits', 'assets', 'market_value']
newdata[cols_to_clean] = newdata[cols_to_clean].replace('[\$,]', '', regex=True)
newdata[cols_to_clean] = newdata[cols_to_clean].replace('[(--)]', np.nan, regex=True).astype(float)
```

```python
newdata
```

```python
fig, (ax1) = plt.subplots(figsize=(5,5))
fig.tight_layout(h_pad=15, w_pad=10)
sns.barplot(data=newdata.sort_values(by='revenues',ascending=False).head(10), x='name', y='revenues',ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title('Top 10 companies based on total revenues')
```

```python
fig, (ax1) = plt.subplots(figsize=(5,5))
fig.tight_layout(h_pad=15, w_pad=10)
sns.barplot(data=newdata.sort_values(by='profits',ascending=False).head(10), x='name', y='profits',ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title('Top 10 companies based on total profits')
```

```python
fig, (ax1) = plt.subplots(figsize=(5,5))
fig.tight_layout(h_pad=15, w_pad=10)
sns.barplot(data=newdata.sort_values(by='market_value',ascending=False).head(10), x='name', y='market_value',ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title('Top 10 companies based on total market_value')
```

```python
newdata['employees'] = pd.to_numeric(newdata['employees'].replace('[\,,]', '', regex=True), errors='coerce')
```

```python
fig, (ax1) = plt.subplots(figsize=(5,5))
fig.tight_layout(h_pad=15, w_pad=10)
sns.barplot(data=newdata.sort_values(by='employees',ascending=False).head(10), x='name', y='employees',ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title('Top 10 companies based on total employees')
```

```python
fig, (ax1) = plt.subplots(figsize=(5,5))
fig.tight_layout(h_pad=15, w_pad=10)
sns.barplot(data=newdata.sort_values(by='assets',ascending=False).head(10), x='name', y='assets',ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)
ax1.set_title('Top 10 companies based on total assets')
```
